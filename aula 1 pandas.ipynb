on_bad_lines {‘error’, ‘warn’, ‘skip’} ou Callable, padrão ‘error’

Especifica o que fazer ao encontrar uma linha inválida (uma linha com campos a mais do que o esperado). Os valores permitidos são:

'error': Levanta uma Exceção (Exception) quando uma linha inválida é encontrada.

'warn': Levanta um aviso (warning) quando uma linha inválida é encontrada e pula (ignora) essa linha.

'skip': Pula as linhas inválidas silenciosamente, sem levantar exceções ou avisos quando são encontradas.

Callable (função chamável): Uma função que processará uma única linha inválida.

Com engine='python': A função deve ter a assinatura (bad_line: list[str]) -> list[str] | None. O parâmetro bad_line é uma lista de strings divididas pelo separador (sep). Se a função retornar None, a linha inválida será ignorada. Se a função retornar uma nova lista de strings com mais elementos do que o esperado, um aviso ParserWarning será emitido e os elementos excedentes serão descartados.

Com engine='pyarrow': A função deve ter a assinatura conforme descrito na documentação do PyArrow: invalid_row_handler.



sepstr, default ‘,’
Character or regex pattern to treat as the delimiter. If sep=None, the C engine cannot automatically detect the separator, but the Python parsing engine can, meaning the latter will be used and automatically detect the separator from only the first valid row of the file by Python’s builtin sniffer tool, csv.Sniffer. In addition, separators longer than 1 character and different from '\s+' will be interpreted as regular expressions and will also force the use of the Python parsing engine. Note that regex delimiters are prone to ignoring quoted data. Regex example: '\r\t'.



```python
#on_bad_lines ignora ou denuncia linhas com erro
df = pd.read_csv("gapminder.csv", on_bad_lines = "skip", sep = ",")

#head limita as linhas
df.head() #mostra apenas as 5 primeiras linhas
```

```python
#rename renomeia as colunas
df = df.rename(columns={"country": "País", "continent": "Continete", "year": "Ano", "lifeExp": "Expectativa de vida", "pop": "População", "gdpPercap": "PIB"})
df.head()
```

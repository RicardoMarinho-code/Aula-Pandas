```python
import pandas as pd
```

```python
df1 = pd.read_excel("Aracaju.xlsx")
df2 = pd.read_excel("Fortaleza.xlsx")
df3 = pd.read_excel("Natal.xlsx")
df4 = pd.read_excel("Recife.xlsx")
df5 = pd.read_excel("Salvador.xlsx")
```

```python
#junta todos os arquivos, um abaixo do outro
df = pd.concat([df1, df2, df3, df4, df5])
```

```python
df.head()
```

```python
df.tail()
```

```python
df.types()
```

```python
#altera o tipo da coluna LojaID
df["LojaID"] = df["LojaID"].astype("object")
```

```python
df.head()
```

```python
#consulta linhas com valores faltantes
df.isnull().sum()
```

```python
#substitui os valores nulos pela média
df["Vendas"].fillna(df["Vendas"].mean(), inplace=True) #inplace: seja modificado em memória
```

```python
#substitui os valores por 0
df["Vendas"].fillna(0, inplace=True) #inplace: seja modificado em memória
```

```python
df.isnull().sum
```

```python
#remove todas as linhas que estejam com valores faltantes em todas as colunas
df.dropna(how="all", inplace=True)
```

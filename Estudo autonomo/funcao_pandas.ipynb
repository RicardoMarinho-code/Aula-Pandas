```python
import pandas as pd

dados = {
    'Nome': ['Ana', 'Bruno', 'Carlos'],
    'Idade': [25, 30, 28],
    'Salário': [3000, 4500, 3800]
}

df = pd.DataFrame(dados)

# Função simples
def salario_medio(dataframe):
    return dataframe['Salário'].mean()

print(df)
print(f"\nSalário médio: R$ {salario_medio(df):.2f}")
```
